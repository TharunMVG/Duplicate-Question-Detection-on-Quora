```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    
    if pd.isnull(text):
        return ''

    text = text.lower()
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    tokens = text.split()
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    clean_text = ' '.join(tokens)

    return clean_text


dff1['question1'] = dff1['question1'].apply(preprocess)
dff1['question2'] = dff1['question2'].apply(preprocess)

dff1.head
```

```python
from nltk.tokenize import word_tokenize
from fuzzywuzzy import fuzz

def extract_basic_features(df):
    dff1['q1_len_char'] = dff1['question1'].apply(len)
    dff1['q2_len_char'] = dff1['question2'].apply(len)
    dff1['q1_len_word'] = dff1['question1'].apply(lambda x: len(word_tokenize(x)))
    dff1['q2_len_word'] = dff1['question2'].apply(lambda x: len(word_tokenize(x)))
    dff1['char_diff'] = abs(dff1['q1_len_char'] - df['q2_len_char'])
    dff1['word_diff'] = abs(dff1['q1_len_word'] - df['q2_len_word'])

    def common_words_count(row):
        q1_words = set(word_tokenize(row['question1']))
        q2_words = set(word_tokenize(row['question2']))
        return len(q1_words.intersection(q2_words))

    dff1['common_words'] = dff1.apply(common_words_count, axis=1)

    dff1['fuzz_ratio'] = dff1.apply(lambda x: fuzz.ratio(x['question1'], x['question2']), axis=1)
    return dff1

dff1= extract_basic_features(dff1)

dff1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

combined_questions = dff1['question1'] + ' ' + dff1['question2']
count_vectorizer = CountVectorizer(tokenizer=lambda x: x.split())
count_matrix = count_vectorizer.fit_transform(combined_questions)

q1_indices = range(len(dff1))
q2_indices = range(len(dff1)) 


jaccard_sim = []
for i, j in zip(q1_indices, q2_indices):
    q1_tokens = set(count_matrix[i].indices)
    q2_tokens = set(count_matrix[j].indices)

    if len(q1_tokens.union(q2_tokens)) == 0:
        similarity = 0
    else:
        similarity = len(q1_tokens.intersection(q2_tokens)) / len(q1_tokens.union(q2_tokens))
    jaccard_sim.append(similarity)

dff1['jaccard_similarity'] = jaccard_sim

dff1
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Bidirectional, LSTM, Dropout, Dense, Concatenate

y=dff1['is_duplicate']

# Separating the text columns and the feature columns
text_columns = ['question1', 'question2']
feature_columns = [
    'q1_len_char', 'q2_len_char', 'q1_len_word', 'q2_len_word',
    'char_diff', 'word_diff', 'common_words', 'fuzz_ratio',
    'cosine_similarity', 'jaccard_similarity'
]

text_data = dff1[text_columns]
feature_data = dff1[feature_columns]
labels = dff1['is_duplicate']

X_text_train, X_text_test, X_feat_train, X_feat_test, y_train, y_test = train_test_split(
    text_data, feature_data, labels, test_size=0.2, random_state=42)


texts = X_text_train['question1'].tolist() + X_text_train['question2'].tolist()
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)

max_length = 100
X_train_seq1 = tokenizer.texts_to_sequences(X_text_train['question1'].tolist())
X_train_seq2 = tokenizer.texts_to_sequences(X_text_train['question2'].tolist())
X_test_seq1 = tokenizer.texts_to_sequences(X_text_test['question1'].tolist())
X_test_seq2 = tokenizer.texts_to_sequences(X_text_test['question2'].tolist())

X_train_pad1 = pad_sequences(X_train_seq1, maxlen=max_length, padding='post')
X_train_pad2 = pad_sequences(X_train_seq2, maxlen=max_length, padding='post')
X_test_pad1 = pad_sequences(X_test_seq1, maxlen=max_length, padding='post')
X_test_pad2 = pad_sequences(X_test_seq2, maxlen=max_length, padding='post')


embedding_dim = 100
vocab_size = 10000

input_q1 = Input(shape=(max_length,), name='input_q1')
input_q2 = Input(shape=(max_length,), name='input_q2')
input_features = Input(shape=(X_feat_train.shape[1],), name='input_features')

embedding_q1 = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length)(input_q1)
lstm_q1 = Bidirectional(LSTM(64, return_sequences=True))(embedding_q1)
lstm_q1 = Bidirectional(LSTM(32))(lstm_q1)

embedding_q2 = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length)(input_q2)
lstm_q2 = Bidirectional(LSTM(64, return_sequences=True))(embedding_q2)
lstm_q2 = Bidirectional(LSTM(32))(lstm_q2)

concat = Concatenate()([lstm_q1, lstm_q2, input_features])

dense = Dropout(0.5)(concat)
dense = Dense(64, activation='relu')(dense)
dense = Dropout(0.5)(dense)
output = Dense(1, activation='sigmoid')(dense)

model = Model(inputs=[input_q1, input_q2, input_features], outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

history = model.fit(
    [X_train_pad1, X_train_pad2, X_feat_train], y_train,
    epochs=10, batch_size=64, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate([X_test_pad1, X_test_pad2, X_feat_test], y_test)
print(f'Test Loss: {loss:.4f}')
print(f'Test Accuracy: {accuracy:.4f}')
predictions = model.predict([X_test_pad1, X_test_pad2, X_feat_test])

# Assuming 'predictions' are probabilities, you can convert them to classes (0 or 1) based on a threshold
threshold = 0.5
predicted_classes = (predictions > threshold).astype(int)
print("Example Predictions:")
for i in range(5):
    print(f"Question 1: {X_text_test.iloc[i]['question1']}")
    print(f"Question 2: {X_text_test.iloc[i]['question2']}")
    print(f"Predicted Duplicate: {predicted_classes[i]}")
    print(f"Actual Duplicate: {y_test.iloc[i]}")
    print()

```
